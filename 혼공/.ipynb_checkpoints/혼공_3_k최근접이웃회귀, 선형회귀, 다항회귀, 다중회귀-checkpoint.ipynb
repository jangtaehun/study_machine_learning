##### 회귀
> - 임의의 어떤 숫자를 예측하는 것
> - 정해진 클래스가 없고 임의의 수치를 출력한다.

##### k-최근접 이웃 회귀
> - 가장 가까운 샘플 k개를 선택(임의의 수치) -> 평균을 이용
> - 회귀에서 정확한 숫자를 맞힌다는 것은 거의 불가능 -> 예측하는 값이나 타깃 모두가 임의의 수치이기 때문
> - 평가 = R^2 = 결정계수 = 클 수록 좋다.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
# print(train_input) -> (42, )
print()

train_input = train_input.reshape(-1, 1) #훈련 세트는 2차원 배열 -> #predict는 2차원 배열을 입력해야 한다.
test_input = test_input.reshape(-1, 1)
#reshape() = 벡터나 배열의 모양을 변경 / 수평에서 수직 = (매개변수로 요소의 개수, 1)
# print(train_input) -> (42, 1)
print()

from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr.fit(train_input, train_target)
print(knr.score(test_input, test_target))
#분류의 경우 정확하게 분류한 개수의 비율 = 정확도 -> 정답을 맞힌 개수의 비율
#회귀의 경우 결정계수 = R^2
#높을 수록 좋다.

from sklearn.metrics import mean_absolute_error #타깃과 예측의 절댓값 오차를 평균하여 반환
#테스트 세트에 대한 예측을 만든다.
test_prediction = knr.predict(test_input)
#테스트 세트에 대한 평균 절댓값 오차를 계산
mae = mean_absolute_error(test_target, test_prediction) # 타깃, 예측값
print(mae) #19g정도 다르다

#훈련 세트에서 점수가 좋지만 테스트 세트에서 점수가 안 좋으면 과대적합
#훈련 세트보다 테스트 세트에 점수가 높거나 / 두 점수가 모두 낮으면 과소적합 => 모델 단순, 훈련X -> 크기가 작다

knr.n_neighbors = 3
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
print(knr.score(test_input, test_target))

knr.predict([[50]])
distance, index = knr.kneighbors([[50]]) #이웃을 구한다
print(distance)
print(index)
plt.scatter(train_input, train_target)
plt.scatter(train_input[index], train_target[index], marker='D')
plt.scatter(50, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print(knr.predict([[100]]))
distance, index = knr.kneighbors([[100]]) #이웃을 구한다
plt.scatter(train_input, train_target)
plt.scatter(train_input[index], train_target[index], marker='D')
plt.scatter(100, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

#훈련세트 다시 만들어서 훈련!!
```

##### 선현회귀
> - 가장 잘 맞는 직선의 방정식을 찾는다. => 최적의 기울기, 절편을 구한다.

```python
#선형회귀
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)

train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

lr = LinearRegression()
lr.fit(train_input, train_target)

lr_predict = lr.predict([[50]])
print(lr_predict)

#coef_ :기울기(계수, 가중치) = theta / intercept_ :y절편 = theta0
plt.scatter(train_input, train_target)
plt.plot([15, 50], [15*lr.coef_ + lr.intercept_, 50*lr.coef_ + lr.intercept_]) #15~50까지 직선 / 농어 무게 = a * 농어 길이 + b
plt.scatter(50, lr_predict, marker = '^')
plt.show()

print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
#다항 회귀 -> 2차원 이상
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, test_size=0.2, random_state=42)
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)


#2차 방정식 그래프 -> 길이 제곱 : train_input 제곱 train_input을 붙인다.
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
print(train_poly.shape)
print(test_poly.shape)
print()

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.predict([[50**2, 50]]))
print(lr.coef_)
print(lr.intercept_)

point = np.arange(15, 50)
plt.scatter(train_input, train_target)
plt.plot(point, 1.01*point**2 - 21.6*point+116.05) #(theta1 * x1) + (theta2 * x2) + theta0
plt.scatter(50, 1574, marker = '^')

print()
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

```python
#다중 회귀 - 여러 개의 특성 사용
# 기존 특성을 사용해 새로운 특성을 뽑아내는 작업을 특성 공학
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#길이 높이 두께
df = pd.read_csv('https://bit.ly/perch_csv_data')
perch_full = df.to_numpy()

#target
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state=42)

# 사이킷런은 특성을 만들거나 전처리하기 위한 다양한 클래스 제공 -> 변환기 => fit, transform 메서드 제공
# 변환기에는 입력 데이터만 전달한다.
# PolynomialFeatures은 주어진 특성을 조합해 새로운 특성을 만든다. degree는 최고 차수를 지정(기본 2)
poly = PolynomialFeatures(include_bias=False) #절편이 1인 값 제외 -> include_bias=False =>(길이, 높이, 두께, 1) ->(길이, 높이, 두께)
train_poly = poly.fit_transform(train_input) #두 개의 특성 2와 3으로 이루어진 샘플 하나를 적용 / fit()을 해야 transform 가능 / ★훈련 세트로 학습한 변환기를 사용해 테스트 세트까지 변환해야 한다.
test_poly = poly.transform(test_input)

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
print()

#특성 추가
poly = PolynomialFeatures(degree=5, include_bias=False) #절편이 1인 값 제외 / 5제곱
poly.fit(train_input) #두 개의 특성 2와 3으로 이루어진 샘플 하나를 적용 / fit()을 해야 transform 가능
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target)) #음수 = 훈련세트 과대적합
print()


#규제 = 훈련 세트를 과도하게 학습하지 못하도록 훼방
# 훈련 세트로 학습한 변환기를 사용해 테스트 세트까지 변환해야 한다.
from sklearn.preprocessing import StandardScaler #평균화
ss = StandardScaler()
train_scaled = ss.fit_transform(train_poly)
test_scaled = ss.transform(test_poly)
#선형 회귀 모델에 규제를 추가하 모델을 릿지, 라쏘
#릿지 = 계수를 제곱한 값을 기준으로 규제 (추천)
#라쏘 = 계수의 절댓값을 기준으로 규제 -> 모든 계수의 크기를 0으로 만들 수 있다.

#릿지 회귀
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))

#alpha 값을 통해 규제 강도 조절, 크면 세진다 -> 과소적합 유도 : 기본값 = 1
#적절한 alpha 찾는 법 => alpha 값에 대한 R^2 값의 그래프 그려보기
train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  ridge = Ridge(alpha = alpha) #릿지 모델 만들기
  ridge.fit(train_scaled, train_target) #훈련
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()

ridge = Ridge(alpha = 0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))

#라쏘 회귀
# 최적 모델을 찾기 위해 좌표축을 따라 최적화를 수행해가는 죄표 하강법을 사용
# max_iter = 알고리즘의 수행 반복 횟수, 기본 1000
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))

train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  lasso = Lasso(alpha = alpha, max_iter=10000) #라쏘 모델 만들기, max_iter 반복 횟수
  lasso.fit(train_scaled, train_target) #훈련
  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()

lasso = Lasso(alpha = 10)
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))

#lasso모델은 계수 값을 0으로 만들 수 있다. -> 유용한 특성을 골라내는데 사용
print(np.sum(lasso.coef_ == 0))
```
