> - 연령대가 넓게 분포해서 정확히 파악하기 어렵다는 것을 알 수 있다.

```python
# 입력 age에 따라 구분값을 반환하는 함수 설정. DataFrame의 apply lambda식에 사용. 
def get_category(age):
    cat = ''
    if age <= -1: cat = 'Unknown'
    elif age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat = 'Elderly'
               
    return cat

#X축의 값을 순차적으로 표시하기 위한 설정 
group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Elderly']
 
# get_category(X)는 입력값으로 'Age' 컬럼값을 받아서 해당하는 cat 반환
train_df['Age_range'] = train_df['Age'].apply(lambda x : get_category(x))
print(train_df[['Age', 'Age_range']])

plt.figure(figsize=(10,6))
sns.barplot(x='Age_range', y = 'Survived', hue='Sex', data=train_df, order=group_names)
```

```python
group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Elderly']
group_survived = {}

for i in group_names:
    group_total = train_df[(train_df['Age_range'] == i)].shape[0]
    group_1 = train_df[(train_df['Age_range'] == i) & (train_df['Survived'] == 1)].shape[0]
    if group_1 != 0:
        group_survived[i] = round(group_1 / group_total * 100, 2) 
    else:
        group_survived[i] = 0


df_group_survived = pd.DataFrame.from_dict(group_survived, orient='index', columns=['Survival Rate (%)'])
df_group_survived = df_group_survived.reset_index()
df_group_survived.columns = ['Age Group', 'Survival Rate (%)']

custom_palette = ["#FF6B6B", "#FFD93D", "#9BDE7C", "#6495ED", "#9370DB", "#E6E6FA", "#FFA07A", "#AFEEEE"]
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Age Group', y='Survival Rate (%)', data=df_group_survived, palette=custom_palette)

# 데이터 레이블 추가 (i = index, v = value)
for i, v in enumerate(df_group_survived['Survival Rate (%)']):
    ax.text(i, v, f"{v:.0f}%", color='black', ha='center', va='bottom', fontsize=10)

# 제목, x축 레이블, y축 레이블 추가
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate (%)')
plt.legend()
plt.show()
```

```python
print('train_df: 데이터 세트 Null 값 갯수 ',train_df.isnull().sum().sum())
train_df.isnull().sum()
```

```python
def drop_features(df):
    df.drop(['PassengerId','Name','Ticket', 'Age', 'Fare'], axis=1, inplace=True)
    print(df.head(3), "\n\n")

    y = df['Survived']
    df = df.drop('Survived', axis=1, inplace=False)
    return df, y


def fillna(df):
    df['Cabin'] = df['Cabin'].fillna('N')
    df['Embarked'] = df['Embarked'].fillna('N')
    # imputer = SimpleImputer(missing_values=np.nan, strategy='mean') #missing_values=np.nan 모든 결측값을 대체한다. / strategy='mean' 평균으로 대체한다.
    # age_array = df['Age'].to_numpy().reshape(-1, 1)
    # imputer.fit(age_array)
    # df['Age'] = imputer.transform(age_array)
    return df
```

```python
train_df = fillna(train_df)
X, y = drop_features(train_df)
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

X = X.values

# 원핫 인코딩 = 한 특성이 여러 개의 범주를 가질 때
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1, 4, 5, 6])], remainder='passthrough')
X = ct.fit_transform(X)

# 표준화 => 2차원 데이터
sc = StandardScaler()
X[:, :] = sc.fit_transform(X[:, :])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    
    accuracy = accuracy_score(y_test , pred)
    
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    
    f1 = f1_score(y_test,pred)
    
    # ROC-AUC 추가 
    roc_auc = roc_auc_score(y_test, pred_proba)
    
    print('오차 행렬')
    print(confusion)
    
    # ROC-AUC print 추가
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))

def precision_recall_curve_plot(y_test=None, pred_proba_c1=None):
    
    # threshold ndarray와 이 threshold에 따른 정밀도, 재현율 ndarray 추출. 
    precisions, recalls, thresholds = precision_recall_curve( y_test, pred_proba_c1)
    
    # X축을 threshold값으로, Y축은 정밀도, 재현율 값으로 각각 Plot 수행. 정밀도는 점선으로 표시
    plt.figure(figsize=(8,6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
    
    # threshold 값 X 축의 Scale을 0.1 단위로 변경
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    
    # x축, y축 label과 legend, 그리고 grid 설정
    plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
    plt.legend(); plt.grid()
    plt.show()
```

# 로지스틱 회귀
### 분류 모델
> - 0~1 사이의 값 => 시그모이드 함수 = 로지스틱 함수 사용
> - 0.5 보다 크면 양성 클래스, 0.5 보다 작으면 음성 클래스

> - 다중분류는 소프트맥스 함수를 사용해 클래스마다 z값을 확률로 변환 / 이진 분류는 시그모이드 함수 사용
> - 소프트맥스 함수 = 여러 개의 선형 방정식의 출력값을 0~1 사이로 압축하고 전체 합이 1이 되도록 만든다.
> - multi_class='multinomial'옵션을 주면 소프트맥스 회귀
> - solver='lbfgs'옵션을 주어 소프트맥스 회귀를 지원하도록 알고리즘을 지정해야 한다.

> - 반복적인 알고리즘 사용 -> max_iter 매개변수로 반복 횟수 지정, default=100
> - 릿지와 같이 계수의 제곱을 규제 = L2 규제
> - 릿지에서는 alpha 매개변수로 규제를 제어, alpha가 커지면 규제도 커진다.
> - 로지스틱 회귀에서는 규제를 제어하는 매개변수는 C -> 값이 작을 수록 규제가 커진다. default=1

#### 소프트맥스
> - lr = LogisticRegression(C=20, max_iter=100, multi_class="multinomial",solver="lbfgs")
> - lr.fit(X_train, y_train)

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

print(lr.score(X_train, y_train)) 
# 훈련 데이터 X_train과 y_train에 대한 정확도 점수
# 정확도 점수는 0과 1 사이의 값을 가지며, 1에 가까울수록 모델의 성능이 좋다는 것을 의미
# 모델이 훈련 데이터에 대해 얼마나 잘 예측하는지 확인

print(lr.score(X_test, y_test))
# 테스트 데이터 X_test와 y_test에 대한 정확도 점수
# 테스트 데이터는 모델 학습에 사용되지 않은 데이터이므로, 이 점수는 모델의 일반화 성능을 나타낸다.

# 훈련 데이터에 대한 정확도와 테스트 데이터에 대한 정확도를 비교하면, 모델의 과적합(overfitting) 여부를 확인할 수 있다.
# 훈련 데이터 정확도가 높고 테스트 데이터 정확도가 낮다면, 모델이 과적합되었을 가능성
# 모델의 성능과 일반화 능력을 평가
```

```python
lr_pred = lr.predict(X_test) 
print('LogisticRegression 정확도: {0:.4f}'.format(accuracy_score(y_test, lr_pred))) 
# 테스트 데이터에 대한 정확도 점수를 계산 = lr.score(X_test, y_test) = lr이 테스트 데이터에 대한 정확도
# lr_pred: 로지스틱 회귀 모델 lr이 테스트 데이터에 대해 예측한 라벨 값
# y_test: 테스트 데이터의 실제 라벨 값

pred_proba = lr.predict_proba(X_test)[:, 1] # 분류 결정 예측 확률
print(pred_proba[:4])
# 로지스틱 회귀 모델 lr이 테스트 데이터 X_test에 대해 예측한 클래스 확률 값을 반환하는 것
# 이진 분류 문제에서는 [[0.3, 0.7], [0.6, 0.4], ...]와 같은 형태로 반환
# 첫 번째 열은 클래스 0의 확률, 두 번째 열은 클래스 1의 확률
# 클래스 확률 값은 모델이 각 데이터 포인트를 어느 클래스로 분류할지 결정하는 데 사용됩니다. 일반적으로 확률 값이 0.5 이상이면 해당 클래스로 분류

get_clf_eval(y_test , lr_pred, pred_proba)
```

# K-Nearest-Neighbors
> - 비모수 기계 학습 알고리즘 중 하나로, 분류와 회귀 문제에 모두 사용

### 작동 원리
> - 데이터 포인트와 그 레이블(분류) 또는 값(회귀)을 저장
> - 새로운 데이터 포인트가 주어지면, 이 포인트와 가장 가까운 K개의 이웃을 찾는다.
> - 분류 문제의 경우, K개의 이웃 중 가장 많은 레이블을 해당 데이터 포인트의 예측 레이블로 사용
> - 회귀 문제의 경우, K개의 이웃의 값들의 평균을 해당 데이터 포인트의 예측값으로 사용

### 특징
> - 비모수 알고리즘: 데이터 분포에 대한 가정이 필요하지 않다.
> - 메모리 기반: 모든 학습 데이터를 저장하고 활용 => 장점이자 단점!! (데이터 크기가 클 경우 예측 속도가 느려질 수 있다는 단점)
> - 지역적 예측: 새로운 데이터 포인트와 가까운 이웃들만을 활용하여 예측
> - 단순하지만 강력한 성능: 복잡한 비선형 문제에도 잘 작동

### 하이퍼파라미터
> - K: 이웃 수를 결정, K가 작을수록 과적합, 클수록 과소적합 위험
> - 거리 척도: 데이터 포인트 간 거리를 측정하는 방식을 결정, 일반적으로 유클리드 거리 많이 사용
> - 가중치: 이웃들의 영향력을 거리에 따라 다르게 줄 수 있다.

#### KNN은 임계값 설정이 필요 없다.
> - 입력 데이터와 가장 가까운 k개의 이웃을 찾아 다수결로 클래스를 결정하는 알고리즘
> - k 값(이웃의 개수)과 거리 측정 방식(metric)과 같은 하이퍼파라미터가 중요한 역할

#### 로지스틱 회귀, SVM, 신경망 등의 다른 분류기에서는 예측 확률이 중요하기 때문에, 임계값 설정이 필요

```python
from sklearn.neighbors import KNeighborsClassifier

knr = KNeighborsClassifier()
knr.fit(X_train, y_train)

print(f"train score: {knr.score(X_train, y_train)}")
print(f"test score: {knr.score(X_test, y_test)}")

knr_y_pred = knr.predict(X_test)
knr_pred_proba = knr.predict_proba(X_test)[:, 1]

get_clf_eval(y_test , knr_y_pred, knr_pred_proba)
```

```python
def KNs(n, metric, weight):
    knr = KNeighborsClassifier(n_neighbors=n, metric=metric, weights= weight)
    knr.fit(X_train, y_train)
    print(f"train score: {knr.score(X_train, y_train)}")
    print(f"test score: {knr.score(X_test, y_test)}")
    knr_y_pred = knr.predict(X_test)
    knr_pred_proba = knr.predict_proba(X_test)[:, 1]
    get_clf_eval(y_test , knr_y_pred, knr_pred_proba)
    print("---------------------------------------------")

n = [3, 5, 7, 9, 11]
metric = ['euclidean', 'manhattan', 'minkowski', 'cosine']
weight = ['uniform', 'distance']

combinations = [(n_neighbors, metric_type, weight_type) for n_neighbors in n for metric_type in metric for weight_type in weight]
for n_neighbors, metric_type, weight_type in combinations:
    print(f"n_neighbors: {n_neighbors}, metric: {metric_type}, weight: {weight_type}")
    KNs(n_neighbors, metric_type, weight_type)
```

```python
# 그리드 서치 수행
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'metric': ['euclidean', 'manhattan', 'minkowski', 'cosine'],
    'weights': ['uniform', 'distance']
}

knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5) # 그리드 서치에서 모델의 성능을 평가하는 지표로 정확도(Accuracy)를 사용
knn_grid = grid_search.fit(X_train, y_train)


# 최적의 하이퍼 파라미터 출력
print("Best Parameters:", knn_grid.best_params_, "\n")

# 훈련 데이터 성능 평가
print("Train Accuracy:", knn_grid.score(X_train, y_train), "\n")

# 테스트 데이터 성능 평가 
print("Test Accuracy:", knn_grid.score(X_test, y_test))
```

```python
knn_best_params = knn_grid.best_estimator_
print(f"train score: {knn_best_params.score(X_train, y_train)}")
print(f"test score: {knn_best_params.score(X_test, y_test)}")
```

```python
y_pred = knn_best_params.predict(X_test)
pred_proba = knn_best_params.predict_proba(X_test)[:, 1]

get_clf_eval(y_test , y_pred, pred_proba)
```

## 그리드 서치와 직접 모든 조합을 테스트하는 방식의 결과 값이 다른 이유
> - 그리드 서치의 교차 검증(Cross-Validation)
>   - 그리드 서치는 교차 검증을 통해 하이퍼 파라미터를 최적화.
>   - 이 과정에서 훈련 데이터를 여러 번 나누어 학습하고 평가하므로, 최종 테스트 데이터 성능과는 차이가 날 수 있다.
> - 데이터 분할의 차이:
>   - 그리드 서치와 직접 테스트하는 경우에 사용한 데이터 분할이 다를 수 있다. 데이터 분할에 따라 모델의 성능이 달라질 수 있다.
> - 하이퍼 파라미터 튜닝 방식의 차이:
>   - 그리드 서치는 전체 하이퍼 파라미터 공간을 체계적으로 탐색하여 최적의 조합을 찾는다. 직접 테스트하는 경우 모든 조합을 확인하지만 각 조합에 대한 최적화가 이루어지지 않는다.
> 그리드 서치와 직접 테스트하는 방식의 결과가 다를 수 있다. 그리드 서치의 결과가 일반적으로 더 안정적이며, 최적의 하이퍼 파라미터를 찾는 데 도움이 된다.
